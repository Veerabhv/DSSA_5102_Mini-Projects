Created the elasticsearch instance to which index data can be saved.

```python
class TweetStreamListener(StreamListener):
# Setting StreamListener to be our listener class.
    
    def on_data(self, data):

        dict_data = json.loads(data)
        print(dict_data)

        # collecting all desired data fields 
        es.index(index="tweetindex",
                 doc_type="test-type",
                 body={"author": dict_data["user"]["screen_name"],
                       "date": dict_data["created_at"],
                       "message": dict_data["text"]
                       })
        return True

    def on_error(self, status):
        print(status)
        
# Check before executing the module
if __name__ == '__main__':

    listener = TweetStreamListener()
    # creating an instance of the tweepy tweet stream listener

    auth = OAuthHandler(ckey, csecret)
    auth.set_access_token(atoken, asecret)
    # Authorizing with twitter keys/tokens

    stream = Stream(auth, listener)
    # creating the tweepy stream

    stream.filter(track=['cricket'])
    # Adding a filter to search the tweets for keyword "cricket" 
```
